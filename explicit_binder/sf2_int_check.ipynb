```python
N_A = 5
N_P = 6.3
b_A = 2.3
b_P = 3
k = np.arange(0,1,.1)
n_i = 6
```

## int 1

```python
#mathematica output
int1_raw = np.exp(-(1/6) * k**2 * (2*N_A*b_A**2 + n_i*b_P**2 )) \
            * (-1 + np.exp((1/6)*b_A**2*N_A*k**2)) \
            * (np.exp((1/6)* b_A**2 *N_A * k**2) - np.exp((1/6)*k**2*( b_P**2 * n_i + b_A**2*N_A)))
```

```python
#simplified
int1_simp = (1-np.exp(-(1/6)*b_A**2*N_A*k**2)) * (-1 + np.exp(-(1/6)*k**2*(b_P**2*n_i)))
```

```python
#andys guess
int1_andy = (1-np.exp(-(1/6)*b_A**2*N_A*k**2)) * (1 - np.exp(-(1/6)*k**2*(b_P**2*n_i)))
```

```python
plt.scatter(k, int1_raw)
plt.scatter(k, int1_simp, marker = "x")
plt.scatter(k, int1_andy, marker = "v")
```
